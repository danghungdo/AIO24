```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Ex II.1.a
def compute_vector_length(vector):
    len_of_vector = np.sqrt(np.sum(np.square(vector)))
    return len_of_vector
```

```python
# Ex II.1.b
def compute_dot_product(vector1, vector2):
    dot_product = np.dot(vector1, vector2)
    return dot_product
```

```python
# Ex II.1.c
def compute_multi_vector(matrix, vector):
    result = np.dot(matrix, vector)
    return result
```

```python
# Ex II.1.d
def compute_multi_matrix(matrix1, matrix2):
    result = np.dot(matrix1, matrix2)
    return result
```

```python
# Ex II.1.e
def inverse_matrix(matrix):
    result = np.linalg.inv(matrix)
    return result
```

### Matrix Inverse Calculation

Matrix A:
$$
A = \begin{pmatrix}
-2 & 6 \\
8 & -4
\end{pmatrix}
$$

1. **Determinant of A**:

The determinant of A is:

$$
\text{det}(A) = (-2 \cdot -4) - (6 \cdot 8) = 8 - 48 = -40
$$
2. **Check if A is invertible**:

$$
\text{det}(A) = -40 \neq 0\
$$
Therefore, matrix A is invertible.

3. **Inverse of A**:

The inverse of A is:
$$
A^{-1} = \frac{1}{\text{det}(A)} \begin{pmatrix}
-4 & -6 \\
-8 & -2
\end{pmatrix} = \frac{1}{-40} \begin{pmatrix}
-4 & -6 \\
-8 & -2
\end{pmatrix} = \begin{pmatrix}
\frac{1}{10} & \frac{3}{20} \\
\frac{1}{5} & \frac{1}{20}
\end{pmatrix}
$$

```python
A = np.array([[-2, 6], [8, -4]])
A_inv = inverse_matrix(A)
print(A_inv)
```

```python
# Ex II.2
def compute_eigenvalues_eigenvectors(matrix):
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    return eigenvalues, eigenvectors
```

### Eigenvector and Eigenvalue Calculation

Given matrix A:
$$
A = \begin{pmatrix}
0.9 & 0.2 \\
0.1 & 0.8
\end{pmatrix}
$$

1. **Eigenvalues**:

The eigenvalues are found by solving the characteristic equation:

$$
\text{det}(A - \lambda I) = 0
$$

For our matrix A:

$$
A - \lambda I = \begin{pmatrix}
0.9 - \lambda & 0.2 \\
0.1 & 0.8 - \lambda
\end{pmatrix}
$$

$$
\text{det}(A - \lambda I) = (0.9 - \lambda)(0.8 - \lambda) - (0.2 \cdot 0.1)
$$

$$
\text{det}(A - \lambda I) = \lambda^2 - 1.7\lambda + 0.7 = 0 
$$
Solve above equation to obtain the eigenvalues:
$$
\lambda = 1 \quad \text{and} \quad \lambda = 0.7 
$$
2. **Eigenvectors**:

For $\lambda = 1$:
$$
(A - I)\mathbf{v} = 0
$$
$$
\begin{pmatrix}
-0.1 & 0.2 \\
0.1 & -0.2
\end{pmatrix} 
\begin{pmatrix}
v_1 \\
v_2
\end{pmatrix} = 0
$$

This gives the eigenvector:
$$
\mathbf{v_1} = \begin{pmatrix}
2 \\
1
\end{pmatrix}
$$

Normalize $\mathbf{v_1}$:
$$
\mathbf{v_1} = \frac{1}{\sqrt{2^2 + 1^2}} \begin{pmatrix}
2 \\
1
\end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix}
2 \\
1
\end{pmatrix}
$$

$$
\mathbf{v_1} = \begin{pmatrix}
\frac{2}{\sqrt{5}} \\
\frac{1}{\sqrt{5}}
\end{pmatrix}
$$

For $\lambda = 0.7$:

$$
(A - 0.7I)\mathbf{v} = 0
$$
$$
\begin{pmatrix}
0.2 & 0.2 \\
0.1 & 0.1
\end{pmatrix} \begin{pmatrix}
v_1 \\
v_2
\end{pmatrix} = 0
$$

This gives the eigenvector:
$$
\mathbf{v_2} = \begin{pmatrix}
-1 \\
1
\end{pmatrix}
$$

Normalize $\mathbf{v_2}$:
$$
\mathbf{v_2} = \frac{1}{\sqrt{(-1)^2 + 1^2}} \begin{pmatrix}
-1 \\
1
\end{pmatrix} = \frac{1}{\sqrt{2}} \begin{pmatrix}
-1 \\
1
\end{pmatrix}
$$
$$
\mathbf{v_2} = \begin{pmatrix}
\frac{-1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}}
\end{pmatrix}
$$


```python
A = np.array([[0.9, 0.2], [0.1, 0.8]])
eigenvectors, eigenvalues = compute_eigenvalues_eigenvectors(A)
print(eigenvectors)
print(eigenvalues)
```

```python
# Ex II.3
def compute_cosine(v1, v2):
    cos_sim = np.dot(v1, v2) / (compute_vector_length(v1) * compute_vector_length(v2))
    return cos_sim
```

### Cosine Similarity Calculation

Given vectors:
$$
\mathbf{x} = \begin{pmatrix}
1 \\
2 \\
3 \\
4
\end{pmatrix}, \quad
\mathbf{y} = \begin{pmatrix}
1 \\
0 \\
3 \\
0
\end{pmatrix}
$$

1. **Dot product**:
$$
\mathbf{x} \cdot \mathbf{y} = (1 \cdot 1) + (2 \cdot 0) + (3 \cdot 3) + (4 \cdot 0) = 1 + 0 + 9 + 0 = 10
$$
2. **Magnitude**:
$$
\| \mathbf{x} \| = \sqrt{1^2 + 2^2 + 3^2 + 4^2} = \sqrt{1 + 4 + 9 + 16} = \sqrt{30}
$$
$$
\| \mathbf{y} \| = \sqrt{1^2 + 0^2 + 3^2 + 0^2} = \sqrt{1 + 0 + 9 + 0} = \sqrt{10}
$$

3. **Cosine Similarity**:
$$
\text{cs}(\mathbf{x}, \mathbf{y}) = \frac{\mathbf{x} \cdot \mathbf{y}}{\| \mathbf{x} \| \| \mathbf{y} \|} = \frac{10}{\sqrt{30} \cdot \sqrt{10}} = \frac{10}{\sqrt{300}} = \frac{10}{10\sqrt{3}} = \frac{1}{\sqrt{3}}
$$


```python
x = np.array([1, 2, 3, 4])
y= np.array([1, 0 ,3, 0])

cos_sim = compute_cosine(x, y)
print(cos_sim)
```

```python
# Ex II.4
bg1_image = cv2.imread('GreenBackground.png')
bg1_image = cv2.resize(bg1_image, (678, 381))

ob_image = cv2.imread('Object.png')
ob_image = cv2.resize(ob_image, (678, 381))

bg2_image = cv2.imread('NewBackground.jpg')
bg2_image = cv2.resize(bg2_image, (678, 381))
```

```python
def compute_difference(bg_img, input_img):
    difference_single_channel = np.abs(bg_img - input_img)
    return difference_single_channel

difference_single_channel = compute_difference(bg1_image, ob_image)
plt.imshow(difference_single_channel)
```

```python
def compute_binary_mask(difference_single_channel):
    difference_binary = np.zeros(difference_single_channel.shape)
    difference_binary[difference_single_channel > 50] = 255
    return difference_binary
binary_mask = compute_binary_mask(difference_single_channel)
plt.imshow(binary_mask)
```

```python
def replace_background(bg1_image, bg2_image, ob_image):
    difference_single_channel = compute_difference(bg1_image, ob_image)
    binary_mask = compute_binary_mask(difference_single_channel)
    output = np.where(binary_mask == 255, ob_image, bg2_image)
    return output

output = replace_background(bg1_image, bg2_image, ob_image)
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
```
